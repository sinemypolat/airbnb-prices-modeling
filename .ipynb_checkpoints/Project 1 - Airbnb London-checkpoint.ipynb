```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Decision Tree 
from sklearn import tree
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
# K-nearest Neighbours
from sklearn.neighbors import KNeighborsRegressor
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
```

```python
# Term Document Matrix from text for amenities column
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Wordcloud and gmaps
from wordcloud import WordCloud,STOPWORDS
import gmaps
gmaps.configure(api_key="***yourapikey***")
```

```python
df_listings = pd.read_csv('listings-2.csv')
df_listings = df_listings[['id','name', 'summary','description','house_rules','host_name',
        'host_response_time', 'host_is_superhost', 'host_verifications','host_has_profile_pic',
        'host_identity_verified', 'neighbourhood_cleansed', 'latitude', 'longitude','room_type',
        'accommodates','bathrooms', 'bedrooms', 'beds', 'amenities', 'security_deposit',
        'cleaning_fee', 'guests_included', 'minimum_nights', 'number_of_reviews', 'review_scores_rating',
        'instant_bookable', 'require_guest_profile_picture','require_guest_phone_verification', 'price']]

df_listings.info()
```

```python
df_listings.iloc[:3,:20]
```

```python
print(type(df_listings['amenities']),df_listings['amenities'].dtype)
```

```python
df_listings['amenities'] = df_listings['amenities'].str.replace( '{','')
df_listings['amenities'] = df_listings['amenities'].str.replace( '}','')
df_listings['amenities'] = df_listings['amenities'].str.replace( '"','') 
df_listings['amenities'].head(2)
```

```python
# Term Document Matrix for Amenities
count_vectorizer =  CountVectorizer(tokenizer=lambda x: x.split(','))
amenities = count_vectorizer.fit_transform(df_listings['amenities'])
df_amenities = pd.DataFrame(amenities.toarray(), columns=count_vectorizer.get_feature_names())
df_amenities = df_amenities.drop('',1,inplace=False)
```

```python
type(df_amenities)
```

```python
# To make sure that price column is not null or 0 for each house
print(df_listings['price'].isnull().sum())
df_listings[df_listings['price'] == 0 ]
```

```python
# Convert 'f' and 't' to 0 and 1
columns = ['host_has_profile_pic','host_is_superhost','host_identity_verified','instant_bookable',
           'require_guest_profile_picture','require_guest_phone_verification']

for c in columns:
    df_listings[c] = df_listings[c].replace('f',0,regex=True)
    df_listings[c] = df_listings[c].replace('t',1,regex=True)
```

```python
# Drop off rows having null values for bathrooms, bedrooms, beds
df_listings.dropna(subset=['bathrooms', 'bedrooms', 'beds','review_scores_rating', 
                           'host_is_superhost', 'host_has_profile_pic','host_identity_verified'], inplace=True)
```

```python
# Convert currency type to float and replace $ signs, change NANs with 0
df_listings['price'] = df_listings['price'].replace( '[\$,)]','', regex=True ).astype(float)
df_listings['security_deposit'] = df_listings['security_deposit'].fillna(value=0)
df_listings['security_deposit'] = df_listings['security_deposit'].replace( '[\$,)]','', regex=True ).astype(float)
df_listings['cleaning_fee'] = df_listings['cleaning_fee'].fillna(value=0)
df_listings['cleaning_fee'] = df_listings['cleaning_fee'].replace( '[\$,)]','', regex=True ).astype(float)
```

```python
df_heatmap = df_listings.loc[:54000,['latitude','longitude','price']]
df_heatmap.sample(3)
```

```python
# gmaps heatmap with prices
fig = gmaps.figure()
layer=gmaps.heatmap_layer(
    df_heatmap[['latitude', 'longitude']], weights=df_heatmap['price'],
    max_intensity=100, point_radius=2.5)
fig.add_layer(layer)
fig
```

```python
# Dummies for room_type
room_type_dummies = pd.get_dummies(df_listings_2['room_type'])
df_listings_2 = pd.concat([df_listings_2, room_type_dummies],axis=1)
df_listings_2 = df_listings_2.drop('room_type',1)
```

## Decision Tree Algorithm





```python
y = df_listings_2['price']
```

```python
X = df_listings_2.drop('price',1)
```

```python
X_np = np.array(X)
y_np = np.array(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_np, y_np, test_size=0.15, random_state=0)
```

```python
y_actual = y_test
```

```python
print("Training set size : ", len(X_train))
print("Test set size : ", len(X_test))
```

```python
model = DecisionTreeRegressor(min_samples_split=10, min_samples_leaf=5)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print('Mean Squared Error : ', mean_squared_error(y_actual,y_pred))
print('Root of MSE : ', mean_squared_error(y_actual,y_pred)**(1/2))
print('R2 Score : ', r2_score(y_actual,y_pred))
```

```python
param_grid = {'min_samples_split': [2,5, 10, 100, 1000], 'min_samples_leaf': [2,5, 10, 100, 1000],
              'max_depth':[2,5, 10, 100, 1000]}
model = GridSearchCV(model,param_grid)
model.fit(X_train ,y_train)
model.best_params_
```

```python
X_train2, X_valid, y_train2, y_valid = train_test_split(X_train,y_train, test_size=0.15, random_state=0 )
print(X_train2.shape)
print(y_train2.shape)
print(X_valid.shape)
print(y_valid.shape)
```

```python
model = DecisionTreeRegressor(min_samples_split=10, min_samples_leaf=10, max_depth=100)
model.fit(X_train2, y_train2)
y_pred2= model.predict(X_valid)
print('Mean Squared Error : ', mean_squared_error(y_valid,y_pred2))
print('Root of MSE : ', mean_squared_error(y_valid,y_pred2)**(1/2))
print('R2 Score : ', r2_score(y_valid,y_pred2))
```

```python
X_train3 = np.concatenate([X_train2,X_valid])
y_train3 = np.concatenate([y_train2,y_valid])
print(X_train3.shape)
print(y_train3.shape)
```

```python
model.fit(X_train3,y_train3)
y_pred3 = model.predict(X_test)
print('Mean Squared Error : ', mean_squared_error(y_test,y_pred3))
print('Root of MSE : ', mean_squared_error(y_test,y_pred3)**(1/2))
print('R2 Score : ', r2_score(y_test,y_pred3))
```

```python
plt.scatter(y_test, y_pred3,marker='.')
plt.show()
```

```python
features = sorted(list(zip(X.columns, model.feature_importances_)), key=lambda x:x[1], reverse=True)
df_features = pd.DataFrame(features)
df_features.head()
```

## K-Nearest Neighbours Algorithm

```python
model = KNeighborsRegressor()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print('Mean Squared Error : ', mean_squared_error(y_actual,y_pred))
print('Root of MSE : ', mean_squared_error(y_actual,y_pred)**(1/2))
print('R2 Score : ', r2_score(y_actual,y_pred))
```

```python
param_grid = {'n_neighbors': [3,5,7], 'weights': ['uniform','distance']}
model = GridSearchCV(model,param_grid)
model.fit(X_train ,y_train)
model.best_params_
```

```python
X_train2, X_valid, y_train2, y_valid = train_test_split(X_train,y_train, test_size=0.15, random_state=0 )
```

```python
model = KNeighborsRegressor(n_neighbors=7, weights='distance')
model.fit(X_train2, y_train2)
y_pred2= model.predict(X_valid)
print('Mean Squared Error : ', mean_squared_error(y_valid,y_pred2))
print('Root of MSE : ', mean_squared_error(y_valid,y_pred2)**(1/2))
print('R2 Score : ', r2_score(y_valid,y_pred2))
```

```python
X_train3 = np.concatenate([X_train2,X_valid])
y_train3 = np.concatenate([y_train2,y_valid])
```

```python
model.fit(X_train3,y_train3)
y_pred3 = model.predict(X_test)
print('Mean Squared Error : ', mean_squared_error(y_test,y_pred3))
print('Root of MSE : ', mean_squared_error(y_test,y_pred3)**(1/2))
print('R2 Score : ', r2_score(y_test,y_pred3))
```

```python
plt.scatter(y_test, y_pred3,marker='.')
plt.show()
```

## Linear Regression Algorithm

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print('Mean Squared Error : ', mean_squared_error(y_actual,y_pred))
print('Root of MSE : ', mean_squared_error(y_actual,y_pred)**(1/2))
print('R2 Score : ', r2_score(y_actual,y_pred))
```

```python
param_grid = {'fit_intercept': [True,False], 'normalize': [True,False]}
model = GridSearchCV(model,param_grid)
model.fit(X_train ,y_train)
model.best_params_
```

```python
X_train2, X_valid, y_train2, y_valid = train_test_split(X_train,y_train, test_size=0.15, random_state=0 )
```

```python
model = LinearRegression(fit_intercept=True , normalize=False)
model.fit(X_train2, y_train2)
y_pred2= model.predict(X_valid)
print('Mean Squared Error : ', mean_squared_error(y_valid,y_pred2))
print('Root of MSE : ', mean_squared_error(y_valid,y_pred2)**(1/2))
print('R2 Score : ', r2_score(y_valid,y_pred2))
```

```python
X_train3 = np.concatenate([X_train2,X_valid])
y_train3 = np.concatenate([y_train2,y_valid])
```

```python
model.fit(X_train3,y_train3)
y_pred3 = model.predict(X_test)
print('Mean Squared Error : ', mean_squared_error(y_test,y_pred3))
print('Root of MSE : ', mean_squared_error(y_test,y_pred3)**(1/2))
print('R2 Score : ', r2_score(y_test,y_pred3))
```

```python
plt.scatter(y_test, y_pred3,marker='.')
plt.show()
```
